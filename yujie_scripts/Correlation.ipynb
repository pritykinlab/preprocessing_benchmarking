```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the sheets into DataFrames
df1 = pd.read_csv('/Genomics/pritykinlab/yujie/preprocessing_benchmarking/snakemake_results/preprocessing_scores/all.tsv', sep='\t')  
df2 = pd.read_csv('/Genomics/pritykinlab/yujie/preprocessing_benchmarking/new_results/aggregated_results.tsv', sep='\t')  
```

```python
df1
```

```python
df2
```

```python
# Rename 'neighbors_count' in df2 to avoid column name collision during merge
df2.rename(columns={'neighbors_count': 'neighbors_count_new_pipeline'}, inplace=True)

df2.rename(columns={'num_hvg': 'num_HVG'}, inplace=True)
df2.rename(columns={'hvg_method': 'HVG_method'}, inplace=True)
```

```python
df2
```

```python
print("Columns in df1:", df1.columns.tolist())
print("Columns in df2:", df2.columns.tolist())
```

```python
# Step 2: Merge the DataFrames on the matching columns
merge_cols = ["num_PCs", "label", "HVG_method", "num_HVG", "norm_method"]
merged_df = pd.merge(df1, df2, on=merge_cols)
```

```python
merged_df
```

```python
# Step 3: Plot the correlation
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['neighbors_count'], merged_df['neighbors_count_new_pipeline'], alpha=0.6)
plt.title('Correlation of Neighbors Count between snakemake and new pipeline')
plt.xlabel('Neighbors Count (snakemake)')
plt.ylabel('Neighbors Count (new pipeline)')

# Optionally, you can compute the correlation coefficient
correlation_coefficient = merged_df['neighbors_count'].corr(merged_df['neighbors_count_new_pipeline'])
plt.annotate(f'Correlation Coefficient: {correlation_coefficient:.2f}', xy=(0.05, 0.95), xycoords='axes fraction')

plt.show()
```
