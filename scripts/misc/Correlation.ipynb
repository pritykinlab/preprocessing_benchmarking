## Loop through to check all datasets with 'new_pipeline' 

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Root directories for the old and new pipelines
root_dir_old_pipeline = '/Genomics/pritykinlab/yujie/preprocessing_benchmarking/results/harmonized_perturb_seq/'
root_dir_new_pipeline = '/Genomics/pritykinlab/yujie/preprocessing_benchmarking/results/harmonized_perturb_seq_new_pipeline/'

# Function to find matching dataset directories
def find_matching_datasets(root_old, root_new):
    datasets_old = next(os.walk(root_old))[1]  # List of directories under the old pipeline
    datasets_new = next(os.walk(root_new))[1]  # List of directories under the new pipeline
    return set(datasets_old).intersection(datasets_new)  # Return the intersection of both sets

# Function to process each matched dataset
def process_datasets(datasets, root_old, root_new):
    for dataset in datasets:
        file_path_old = os.path.join(root_old, dataset, 'aggregated_results.tsv')
        file_path_new = os.path.join(root_new, dataset, 'aggregated_results.tsv')

        # Check if both files exist before proceeding
        if os.path.exists(file_path_old) and os.path.exists(file_path_new):
            # Load the sheets into DataFrames
            df1 = pd.read_csv(file_path_old, sep='\t')
            df2 = pd.read_csv(file_path_new, sep='\t')

            # Rename 'neighbors_count' in df2 to avoid column name collision
            df2.rename(columns={'neighbors_count': 'neighbors_count_new_pipeline'}, inplace=True)

            # Merge the DataFrames on the matching columns
            merge_cols = ["label", "hvg_norm_combo", "num_hvg", "num_nn", "num_pcs"]
            merged_df = pd.merge(df1, df2, on=merge_cols)

            # Plot the correlation
            plt.figure(figsize=(10, 6))
            plt.scatter(merged_df['neighbors_count'], merged_df['neighbors_count_new_pipeline'], alpha=0.6)
            title = f'Correlation of Neighbors Count between Old and New Pipeline - {dataset}'
            plt.title(title)
            plt.xlabel('Neighbors Count (old pipeline)')
            plt.ylabel('Neighbors Count (new pipeline)')
            correlation_coefficient = merged_df['neighbors_count'].corr(merged_df['neighbors_count_new_pipeline'])
            plt.annotate(f'Correlation Coefficient: {correlation_coefficient:.2f}', xy=(0.05, 0.95), xycoords='axes fraction')
            plt.show()

# Find matching datasets
matching_datasets = find_matching_datasets(root_dir_old_pipeline, root_dir_new_pipeline)

# Process each dataset
process_datasets(matching_datasets, root_dir_old_pipeline, root_dir_new_pipeline)
```
