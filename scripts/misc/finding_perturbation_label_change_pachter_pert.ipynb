# Edit the GehringPachter Dataset

```python
adata = sc.read_h5ad(os.path.join("/Genomics/pritykinlab/yujie/preprocessing_benchmarking/datasets/harmonized_perturb_datasets",
                                  "GehringPachter2019.h5ad"))
```

```python
adata.obs.columns
```

```python
def get_pert(r):
    return f"dose1_{r['dose_value']}_dose2_{r['dose_value_2']}_dose3_{r['dose_value_3']}_dose4_{r['dose_value_4']}"
adata.obs['perturbation'] = adata.obs.apply(get_pert, axis=1)
```

```python
adata.write_h5ad(os.path.join("/Genomics/pritykinlab/yujie/preprocessing_benchmarking/datasets/harmonized_perturb_datasets",
                                  "GehringPachter2019.h5ad"))
```
