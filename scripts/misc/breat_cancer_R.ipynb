## Read in the Bassez dataset - cohort1

```r
# Read in data
##Read in full single-cell object of cohort 1: 29 primary BC patients (all subtypes) treated with anti-PD1 before surgery, samples taken pre and on-treatment; 175,942 cells total; 5’ scRNAseq + scTCR-seq

bassez_counts <- readRDS("/Genomics/pritykinlab/yujie/preprocessing_benchmarking/datasets/breast_cancer/Bassez/1863-counts_cells_cohort1.rds")
bassez_meta <- read.csv("/Genomics/pritykinlab/yujie/preprocessing_benchmarking/datasets/breast_cancer/Bassez/1872-BIOKEY_metaData_cohort1_web.csv")
rownames(bassez_meta) <- bassez_meta$Cell


# Create Seurat object


bassez <- CreateSeuratObject(counts = bassez_counts, assay = "RNA",
  meta.data = bassez_meta)
```

## Read in the Qian dataset

```r
library(Seurat)


# Read in Qian
## Read in full single-cell object: 14 treatment-naïve primary BCs, 44.000 cells (all cell types, of which 14.400 T cells), 5’ scRNA-seq (10x Genomics)


qian_counts <- Read10X("/Genomics/pritykinlab/yujie/preprocessing_benchmarking/datasets/breast_cancer/qian/export/BC_counts")
qian_meta <- read.csv("/Genomics/pritykinlab/yujie/preprocessing_benchmarking/datasets/breast_cancer/qian/2103-Breastcancer_metadata.csv")
rownames(qian_meta) <- qian_meta$Cell

qian <- CreateSeuratObject(counts = qian_counts, assay = "RNA",
  meta.data = qian_meta)
```
