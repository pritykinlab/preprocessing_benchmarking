```python
links = """
https://cf.10xgenomics.com/samples/cell-vdj/3.1.0/vdj_v1_hs_pbmc3_protein/vdj_v1_hs_pbmc3_protein_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-vdj/3.1.0/vdj_nextgem_hs_pbmc3_protein/vdj_nextgem_hs_pbmc3_protein_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/3.1.0/5k_pbmc_protein_v3/5k_pbmc_protein_v3_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/3.1.0/5k_pbmc_protein_v3_nextgem/5k_pbmc_protein_v3_nextgem_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-vdj/3.0.0/vdj_v1_hs_pbmc2_5gex_protein/vdj_v1_hs_pbmc2_5gex_protein_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/3.0.0/pbmc_10k_protein_v3/pbmc_10k_protein_v3_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/3.0.0/malt_10k_protein_v3/malt_10k_protein_v3_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/3.0.0/pbmc_1k_protein_v3/pbmc_1k_protein_v3_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/3.0.2/5k_pbmc_protein_v3/5k_pbmc_protein_v3_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/3.0.2/5k_pbmc_protein_v3_nextgem/5k_pbmc_protein_v3_nextgem_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/4plex_mouse_LymphNode_Spleen_TotalSeqC_multiplex_LymphNode1_BC1_AB1/4plex_mouse_LymphNode_Spleen_TotalSeqC_multiplex_LymphNode1_BC1_AB1_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/4plex_mouse_LymphNode_Spleen_TotalSeqC_multiplex_LymphNode2_BC2_AB2/4plex_mouse_LymphNode_Spleen_TotalSeqC_multiplex_LymphNode2_BC2_AB2_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/4plex_mouse_LymphNode_Spleen_TotalSeqC_multiplex_Spleen1_BC3_AB3/4plex_mouse_LymphNode_Spleen_TotalSeqC_multiplex_Spleen1_BC3_AB3_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/4plex_mouse_LymphNode_Spleen_TotalSeqC_multiplex_Spleen2_BC4_AB4/4plex_mouse_LymphNode_Spleen_TotalSeqC_multiplex_Spleen2_BC4_AB4_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/4plex_DTC_kidney_lung_breast_TotalSeqC_multiplex_Kidney_Cancer1_BC1_AB1/4plex_DTC_kidney_lung_breast_TotalSeqC_multiplex_Kidney_Cancer1_BC1_AB1_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/4plex_DTC_kidney_lung_breast_TotalSeqC_multiplex_Kidney_Cancer2_BC2_AB2/4plex_DTC_kidney_lung_breast_TotalSeqC_multiplex_Kidney_Cancer2_BC2_AB2_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/4plex_DTC_kidney_lung_breast_TotalSeqC_multiplex_Lung_Cancer_BC4_AB4/4plex_DTC_kidney_lung_breast_TotalSeqC_multiplex_Lung_Cancer_BC4_AB4_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/4plex_DTC_kidney_lung_breast_TotalSeqC_multiplex_Breast_Cancer_BC3_AB3/4plex_DTC_kidney_lung_breast_TotalSeqC_multiplex_Breast_Cancer_BC3_AB3_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_01_BC1_AB1/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_01_BC1_AB1_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_02_BC2_AB2/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_02_BC2_AB2_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_03_BC3_AB3/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_03_BC3_AB3_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_04_BC4_AB4/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_04_BC4_AB4_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_05_BC5_AB5/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_05_BC5_AB5_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_06_BC6_AB6/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_06_BC6_AB6_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_07_BC7_AB7/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_07_BC7_AB7_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_08_BC8_AB8/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_08_BC8_AB8_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_09_BC9_AB9/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_09_BC9_AB9_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_10_BC10_AB10/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_10_BC10_AB10_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_11_BC11_AB11/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_11_BC11_AB11_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_12_BC12_AB12/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_12_BC12_AB12_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_13_BC13_AB13/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_13_BC13_AB13_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_14_BC14_AB14/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_14_BC14_AB14_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_15_BC15_AB15/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_15_BC15_AB15_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-exp/7.2.0/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_16_BC16_AB16/16plex_human_PBMC_TotalSeqC_multiplex_PBMC_16_BC16_AB16_count_sample_filtered_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-vdj/6.1.2/10k_PBMC_TBNK_connect_10k_PBMC_TBNK_connect/10k_PBMC_TBNK_connect_10k_PBMC_TBNK_connect_count_sample_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-vdj/6.1.2/5k_human_antiCMV_T_TBNK_connect_5k_human_antiCMV_T_TBNK_connect/5k_human_antiCMV_T_TBNK_connect_5k_human_antiCMV_T_TBNK_connect_count_sample_feature_bc_matrix.h5
https://cf.10xgenomics.com/samples/cell-vdj/6.1.2/7k_melanoma_TBNK_connect_7k_melanoma_TBNK_connect/7k_melanoma_TBNK_connect_7k_melanoma_TBNK_connect_count_sample_feature_bc_matrix.h5
""".strip().split()

names = []
for link in links:
    names.append(link.split('/')[6])

# Directory to store the data
data_dir = "/Genomics/pritykinlab/yujie/preprocessing_benchmarking/datasets/CITE_seq/10x_CITE_seq"

# Subdirectories for h5 and h5ad files
h5_dir = os.path.join(data_dir, 'h5')
h5ad_dir = os.path.join(data_dir, 'h5ad')

# Create directories if they don't exist
os.makedirs(h5_dir, exist_ok=True)
os.makedirs(h5ad_dir, exist_ok=True)

# Download and process each dataset
for name, link in zip(names, links):
    # Define file paths
    h5_file_path = os.path.join(h5_dir, name + '.h5')
    h5ad_file_path = os.path.join(h5ad_dir, name + '.h5ad')

    # Download the file using wget via os.system
    wget_command = f"wget -O '{h5_file_path}' '{link}'"
    print(f"Downloading {name} using command: {wget_command}")
    os.system(wget_command)

    # Load the dataset into Scanpy
    print(f"Loading {name} into Scanpy...")
    adata = sc.read_10x_h5(h5_file_path, gex_only=False)

    # Save the entire dataset in h5ad format
    print(f"Saving {name} dataset in h5ad format to {h5ad_file_path}...")
    adata.write_h5ad(h5ad_file_path)

print("All datasets have been processed and saved.")
```

```python
# Initialize an empty dictionary to store dataset names and their adata
dataset_adata_dict = {}

# Iterate over the dataset names to load the adata objects
for name in names:
    # Define the path to the h5ad file
    h5ad_file_path = os.path.join(h5ad_dir, name + '.h5ad')

    # Load the adata object from the h5ad file
    print(f"Loading {name} dataset from {h5ad_file_path}...")
    adata = sc.read_h5ad(h5ad_file_path)

    # Add the dataset name and its adata to the dictionary
    dataset_adata_dict[name] = adata
```

```python
for key, adata in dataset_adata_dict.items():
    print("##################################")
    print(f"############ {key} ################")
    print("##################################")
    print(adata.shape)
    print(adata.var['feature_types'].value_counts())
```

```python
# Filter datasets
for key, adata in list(dataset_adata_dict.items()):
    feature_types = adata.var['feature_types'].value_counts()
    if 'Gene Expression' not in feature_types or 'Antibody Capture' not in feature_types:
        print(f"Removing {key} as it does not have both Gene Expression and Antibody Capture.")
        dataset_adata_dict.pop(key)
        os.system(f"rm '{os.path.join(h5ad_dir, key + '.h5ad')}'")
```
